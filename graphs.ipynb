```python
# !pip install matplotlib ipympl
import numpy as np
from matplotlib import pyplot as plt
%matplotlib widget
```

```python
file = './results/ga/ex1_whole[2023-02-27][15-08].txt'
file = './results/ga/ex1_whole_101[2023-02-27][19-00].txt'
```

```python
D = np.loadtxt(file, dtype=np.float32)
gen, pop = D.shape
best = D.argmin()
print(f"population={pop} generations={gen} best@{best//pop}={D.flatten()[best]}")

fig = plt.figure()
_ = { 'axis': 1 }
plt.plot(D.mean(**_), label="mean")
plt.plot(D.min(**_), label="min")
plt.plot(D.max(**_), label="max")
plt.legend()
display(fig)
```
