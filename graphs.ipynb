```python
D = np.loadtxt(file, dtype=np.float64)
def U(i): return np.unravel_index(i, D.shape)  # type: ignore


gen, pop = D.shape
best = U(D.argmin())
worst = U(D.argmax())
print(
    f"population={pop}",
    f"generations={gen}",
    f"best@{best[0]}={D[best]}",
    f"worst@{worst[0]}={D[worst]}",
)

D.sort(axis=1)
R = (pop//5) + (pop%5)
S = D[:,:-R]

fig = plt.figure()
_ = {'axis': 1}
plt.plot(D.mean(**_), label="mean")
plt.plot(D.min(**_), label="min")
plt.plot(D.max(**_), label="max")
plt.plot(S.max(**_), label="s-max")
plt.plot(S.mean(**_), label="s-mean")
# plt.yscale('log')
plt.legend()
display(fig)
```
