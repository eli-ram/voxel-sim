```python
# !pip install matplotlib ipympl
import numpy as np
from matplotlib import pyplot as plt
%matplotlib widget
```

```python
file = './results/ga/ex1_whole[2023-02-27][15-08].txt'
file = './results/ga/ex1_whole_101[2023-02-27][19-00].txt'
file = './results/ga/ex1_whole_101_25[2023-02-27][23-17].txt'
file = './results/ga/ex1_whole_101_25[2023-02-28][08-57].txt'
file = './results/ga/ex1_fracture_P41_R35[2023-02-28][10-39].txt'
file = './results/ga/ex1_whole_P101_R35[2023-03-01][09-01].txt'
```

```python
D = np.loadtxt(file, dtype=np.float64)
def U(i): return np.unravel_index(i, D.shape)  # type: ignore

print((D == 1E10).sum())
D[D == 1E10] = np.nan

gen, pop = D.shape
best = U(D.argmin())
worst = U(D.argmax())
print(
    f"population={pop}",
    f"generations={gen}",
    f"best@{best[0]}={D[best]}",
    f"worst@{worst[0]}={D[worst]}",
)

# D.sort(axis=1)
# S = D[:,5:-5]

fig = plt.figure()
_ = {'axis': 1}
plt.plot(D.mean(**_), label="mean")
plt.plot(D.min(**_), label="min")
plt.plot(D.max(**_), label="max")
# plt.plot(S.mean(**_), label="smooth")
plt.legend()
display(fig)
```
